# univariate analysis

**def**: analying one variable at a time is called univariate analysis. it is the simplest form of analyzing data. it is used to describe the data and find patterns in the data. it is also used to find outliers in the data.

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

# 1. Categorical Data

## a. Countplot

```python
sns.countplot(df['Embarked'])
# df['Survived'].value_counts().plot(kind='bar')
```

## b. PieChart

```python
df['Sex'].value_counts().plot(kind='pie',autopct='%.2f')
```

# 2. Numerical Data

## a. Histogram

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'],bins=5) #distribution of age in the data.
```

## b. Distplot

```python
sns.displot(df['Age'])
```

## c. Boxplot

```python
# use of Boxplot to find outliers in the data.
sns.boxplot(df['Fare'])
```

```python
df['Age'].min()
```

```python
df['Age'].max()
```

```python
df['Age'].mean()
```

```python
df['Age'].skew() # skewness of the data. means the data is not normally distributed.
```
