```python
import pandas as pd
import seaborn as sns
```

```python
# load the data built-in dataset in seaborn library.
tips = sns.load_dataset('tips')
```

```python
titanic = pd.read_csv('train.csv')
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
iris = sns.load_dataset('iris')
```

## 1. Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='sex',
    style='smoker',
    size='size'
)
```

## 2. Bar Plot (Numerical - Categorical)

```python
titanic.head()
```

```python
# sns.barplot(titanic['Pclass'],titanic['Age'],hue=titanic['Sex'])
sns.barplot(data=titanic,x='Pclass',y='Age',hue='Sex')
```

## 3. Box Plot (Numerical - Categorical)

```python
sns.boxplot(data=titanic,x='Sex',y='Age',hue='Survived')
```

## 4. Distplot (Numerical - Categorical)

```python
sns.kdeplot(
    data=titanic[titanic['Survived'] == 0],
    x='Age'
)
sns.kdeplot(
    data=titanic[titanic['Survived'] == 1],
    x='Age'
)
```

```python
sns.histplot(
    data=titanic[titanic['Survived'] == 0],
    x='Age',
    kde=True
)
```

## 5. HeatMap (Categorical - Categorical)

```python
titanic.head(3)
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
(titanic.groupby('Pclass')['Survived'].mean() * 100).plot(kind='bar')
```

## 6. ClusterMap (Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived'])) #to find relationship between two categorical variables.
```

## 7. Pairplot

```python
iris.head()
```

```python
sns.pairplot(iris,hue='species') # to check all numerical variables relationship with each other.
```

## 8. Lineplot (Numerical - Numerical)

```python
flights.head()
```

```python
new = flights.groupby('year').sum().reset_index()
```

```python
sns.lineplot(new['year'],new['passengers'])
```
